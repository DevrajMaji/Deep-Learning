Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

->Deep Learning is a subfield of Machine Learning that uses neural networks with many layers (deep neural networks) to automatically learn patterns and features from large amounts of data, similar to how the human brain works.

Evolution of Deep Learning (Brief)

1950s–60s: Early neural models like the Perceptron were introduced.

1980s: Backpropagation algorithm made training neural networks practical.

1990s–2000s: Limited success due to low computing power and small datasets.

2010s onwards: Explosion of Deep Learning due to:

Large datasets (Big Data)

Powerful GPUs

Improved algorithms (ReLU, CNNs, RNNs)

Difference between Deep Learning and Traditional Machine Learning:

Traditional Machine Learning uses manual feature extraction, while Deep Learning automatically learns features.

Traditional Machine Learning works well with small or medium datasets, while Deep Learning requires large datasets.

Traditional Machine Learning uses shallow models, while Deep Learning uses deep neural networks with many layers.

Traditional Machine Learning depends more on human expertise, while Deep Learning depends more on data and computation.

Traditional Machine Learning has lower computational requirements, while Deep Learning needs high computational power (GPUs).

Traditional Machine Learning is suitable for simple problems, while Deep Learning is suitable for complex problems like image and speech recognition.

Traditional Machine Learning is easier to interpret, while Deep Learning models are harder to interpret.

Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?

->A Perceptron is the simplest neural network model, consisting of:

Input nodes (x₁, x₂, …, xₙ) that receive input features

Weights (w₁, w₂, …, wₙ) associated with each input

Bias (b) to shift the decision boundary

Summation unit that computes weighted sum

Activation function (usually step function) to produce output

Functioning (Working) of a Perceptron

Inputs are multiplied by their corresponding weights.

All weighted inputs are summed.

Bias is added to the sum.

The activation function is applied to decide the output (0 or 1).

Weights are updated during training using the error between predicted and actual output.

Limitations of a Perceptron

Can solve only linearly separable problems.

Fails to learn non-linear patterns (e.g., XOR problem).

Uses binary output only.

Cannot handle complex real-world data.

Single-layer structure limits learning capacity.

Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

->The activation function introduces non-linearity into a neural network, allowing it to learn complex patterns.
Without an activation function, a neural network would behave like a linear model, no matter how many layers it has.

Comparison of Sigmoid, Tanh, and ReLU (Line by Line)

Sigmoid outputs values between 0 and 1, while Tanh outputs values between –1 and +1, and ReLU outputs 0 or positive values.

Sigmoid is mainly used for binary classification, Tanh is used in hidden layers, and ReLU is widely used in deep networks.

Sigmoid suffers from the vanishing gradient problem, Tanh reduces it slightly, and ReLU mostly avoids it.

Sigmoid is not zero-centered, Tanh is zero-centered, and ReLU is partially zero-centered.

Sigmoid is computationally expensive, Tanh is moderately expensive, and ReLU is computationally efficient.

Sigmoid learns slowly in deep networks, Tanh learns faster than Sigmoid, and ReLU enables faster training.

Sigmoid can cause saturation at extreme values, Tanh also saturates, while ReLU may cause dead neurons.

Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.

->A Loss function measures the error for a single training example, while a Cost function measures the average error over the entire dataset.

A Loss function is calculated per data point, while a Cost function is calculated over all training samples.

A Loss function helps evaluate individual predictions, while a Cost function evaluates the overall model performance.

A Loss function is mainly used during weight updates for one sample, while a Cost function is used to optimize the model during training.

A Loss function gives local error, while a Cost function gives global error.

Question 5: What is the role of optimizers in neural networks? Compare Gradient
Descent, Adam, and RMSprop.

->Optimizers are used to update the weights and bias of a neural network in such a way that the loss (or cost) function is minimized.
They control how fast and in which direction the model learns by using gradients.

Comparison of Gradient Descent, Adam, and RMSprop (Line by Line)

Gradient Descent uses a single fixed learning rate, while Adam and RMSprop use adaptive learning rates.

Gradient Descent updates weights using the entire dataset, while Adam and RMSprop work efficiently with mini-batches.

Gradient Descent is slow and sensitive to learning rate, while Adam converges faster, and RMSprop stabilizes training.

Gradient Descent can get stuck in local minima, while Adam handles noisy gradients better, and RMSprop reduces oscillations.

Gradient Descent is simple and easy to understand, while Adam is more advanced, and RMSprop is designed for non-stationary objectives.

Gradient Descent uses only first-order gradients, while Adam uses momentum + RMSprop concepts, and RMSprop uses moving average of squared gradients.

Gradient Descent requires careful tuning, while Adam needs less tuning, and RMSprop needs moderate tuning.



```python
'''Question 6: Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.'''

import numpy as np

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

weights = np.zeros(X.shape[1])
bias = 0
learning_rate = 0.1
epochs = 10

def step_function(z):
    return 1 if z >= 0 else 0

for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        y_pred = step_function(linear_output)
        error = y[i] - y_pred
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

print("Trained weights:", weights)
print("Bias:", bias)

for i in range(len(X)):
    output = step_function(np.dot(X[i], weights) + bias)
    print(X[i], "->", output)
```

```python
'''Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.
'''
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, relu, label="ReLU")
plt.plot(x, tanh, label="Tanh")

plt.legend()
plt.xlabel("x")
plt.ylabel("Activation value")
plt.show()
```

```python
'''Question 8: Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.
'''

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train / 255.0
x_test = x_test / 255.0

y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation="relu"),
    Dense(10, activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

history = model.fit(x_train, y_train, epochs=5, batch_size=32, verbose=1)

print("Training Accuracy:", history.history["accuracy"][-1])
```

```python
'''Question 9: Visualize the loss and accuracy curves for a neural network model trained
on the Fashion MNIST dataset. Interpret the training behavior.
'''
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.datasets import fashion_mnist
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

x_train = x_train / 255.0
x_test = x_test / 255.0

model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation="relu"))
model.add(Dense(10, activation="softmax"))

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history = model.fit(
    x_train, y_train,
    epochs=10,
    validation_data=(x_test, y_test)
)

plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend(["Training Loss", "Validation Loss"])
plt.show()

plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend(["Training Accuracy", "Validation Accuracy"])
plt.show()
```

Question 10: You are working on a project for a bank that wants to automatically detect
fraudulent transactions. The dataset is large, imbalanced, and contains structured
features like transaction amount, merchant ID, and customer location. The goal is to
classify each transaction as fraudulent or legitimate.
Explain your real-time data science workflow:
● How would you design a deep learning model (perceptron or multilayer NN)?
● Which activation function and loss function would you use, and why?
● How would you train and evaluate the model, considering class imbalance?
● Which optimizer would be suitable, and how would you prevent overfitting?

->1. Model Design:

Use a Multilayer Neural Network (MLP) with input layer for transaction features, 2–3 hidden layers to learn complex patterns, and 1 output neuron for fraud classification.

Single-layer perceptron is insufficient due to non-linear relationships in fraud patterns.

2. Activation and Loss Function:

ReLU in hidden layers → fast training, avoids vanishing gradient.

Sigmoid in output → gives probability for binary classification.

Binary Cross-Entropy Loss → measures how close predicted probabilities are to actual labels.

3. Training & Evaluation:

Handle class imbalance using class weights, oversampling (SMOTE), or undersampling.

Train with mini-batches, monitor validation metrics.

Evaluate with precision, recall, F1-score, and ROC-AUC instead of just accuracy.

4. Optimizer & Overfitting Control:

Adam optimizer → adaptive learning rate, fast convergence.

Prevent overfitting with dropout, L2 regularization, batch normalization, and early stopping.
